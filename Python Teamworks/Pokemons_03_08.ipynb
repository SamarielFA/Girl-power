POKEMON EXPLORATORY DATA ANALYSIS

Answer the following questions:

1.
How many Pokémons are with 'Type 1' == Water as a % of total?

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Pokemon.csv')
unique_names = df['Name'].nunique()
total = len(df)
print(f"There are totally: {total } Pokemons with  {unique_names} unique names in the dataset")

water_type1 = df[df['Type 1'] == 'Water']
print(f"number of Type1-Water Pokemons: {len(water_type1)} ")


percentage_water_type1 = (len(water_type1) / total) * 100
print( f"Type1-Water Pokemons makes up: {percentage_water_type1:.2f} % of all")
```

2.
What is the maximum 'Speed' value? What is the minimum 'Speed' value? What is the difference between max and min 'Speed'?


```python
max_speed = df['Speed'].max()
min_speed = df['Speed'].min()
speed_difference = max_speed - min_speed
print(f"Maximum Speed: {max_speed}")
print(f"Minimum Speed: {min_speed}")
print(f"Speed Difference: {speed_difference}")
```


3.
Filter the DataFrame to include only the Pokémon with 'Speed' >= 80. How many Pokémon meet this criterion? Display this DataFrame using your preferred visualization method.

```python
speed_80 = df[df['Speed'] >= 80]
speed_less_80 = df[df['Speed'] < 80]
print(f"Number of Pokémon with Speed  80 and more: {len(speed_80)}")
print(f"Number of Pokémon with Speed less than  80: {len(speed_less_80)}")


labels = ['Speed >= 80\nNumber of Pokémon:  296 ', 'Speed < 80\nNumber of Pokémon: 504']
sizes = [len(speed_80), len(speed_less_80)]
colors = ['skyblue', 'lightgreen']
plt.figure(figsize=(6, 4))
plt.pie(sizes,
        labels=labels,
        colors=colors,
        autopct='%1.1f%%',
        startangle=140,
        textprops={'color': 'darkgreen', 'fontsize': 12, 'weight': 'bold'},
        labeldistance=1.1,
        wedgeprops={'linewidth': 1, 'edgecolor': 'darkgreen'}
        )
plt.legend(labels, loc="upper left", fontsize=8, frameon=True, edgecolor='darkgreen', bbox_to_anchor=(1, 0.2))
plt.title('Distribution of Pokémon Speed', fontsize=18, color='darkblue', fontweight='bold')
plt.axis('equal')

plt.show()
```


4.
(DIFFICULT) Find Pokémon with the longest name (excluding spaces)? What is this pokemons name?

```python
df['Name_Length'] = df['Name'].str.replace(' ', '').str.len()# remove spaces  and calculate length
max_length = df['Name_Length'].max()
longest_name_pokemon = df[df['Name_Length'] == max_length]

for _, row in longest_name_pokemon.iterrows():
  print(f"Pokémon with the longest name: {row['Name']} ")
print(f"Length of the name(excluding spaces): {max_length} letters ")
```
